```python
import networkx as nx
# Buat grafik
G = nx.DiGraph()
```

```python
G.add_edge(1 , 2)
G.add_edge(1 , 6)
G.add_edge(1 , 9)
G.add_edge(2 , 6)
G.add_edge(2 , 8)
G.add_edge(3 , 1)
G.add_edge(3 , 5)
G.add_edge(4 , 1)
G.add_edge(4 , 5)
G.add_edge(5 , 3)
G.add_edge(5 , 7)
G.add_edge(6 , 8)
G.add_edge(7 , 4)
G.add_edge(8 , 3)
G.add_edge(9 , 6)
```

```python
nx.draw(G, with_labels=True, arrows=True)
```

```python
# Hitung PageRank
pagerank_values = nx.pagerank(G, alpha=0.85)

# Urutkan hasil PageRank berdasarkan nomor node
sorted_pagerank = sorted(pagerank_values.items())

# Tampilkan hasil yang sudah diurutkan
for node, value in sorted_pagerank:
    print(f"Node {node}: {value}")
```
