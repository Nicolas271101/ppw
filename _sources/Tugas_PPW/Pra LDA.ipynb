```python
df["abstrak"] = df['abstrak'].str.lower()
df["abstrak"]
```

```python
#cleaning
import re
clean =[]
for i in range (len(df["abstrak"])):
  clean_tag  = re.sub("@[A-Za-z0-9_]+","", df["abstrak"][i]) #clenasing mention
  clean_hashtag = re.sub("#[A-Za-z0-9_]+","", clean_tag) #clenasing hashtag
  clean_https = re.sub(r'http\S+', '', clean_hashtag) #cleansing url link
  clean_symbols = re.sub("[^a-zA-Zï ]+"," ", clean_https) #cleansing character

  clean.append(clean_symbols)


df["abstrak"] = clean
df["abstrak"]
```

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
```

```python
# Membuat kamus slang words dan kata Indonesia yang benar
slang_dict = pd.read_csv("https://raw.githubusercontent.com/Nicolas271101/ppw/main/combined_slang_words.txt", sep=" ", header=None)

# Membuat fungsi untuk mengubah slang words menjadi kata Indonesia yang benar
def replace_slang_words(text):
    words = nltk.word_tokenize(text.lower())
    words_filtered = [word for word in words if word not in stopwords.words('indonesian')]
    for i in range(len(words_filtered)):
        if words_filtered[i] in slang_dict:
            words_filtered[i] = slang_dict[words_filtered[i]]
    return ' '.join(words_filtered)

# Contoh penggunaan

slang_words=[]
for i in range(len(clean)):
  slang = replace_slang_words(clean[i])
  slang_words.append(slang)

df["abstrak"] = slang_words
df["abstrak"]
```

```python
#proses tokenisasi
# from nltk.tokenize import TweetTokenizer
# def word_tokenize(text):
#   tokenizer = TweetTokenizer(preserve_case=False, strip_handles=True, reduce_len=True)
#   return tokenizer.tokenize(text)

df["abstrak"]= df["abstrak"].apply(lambda sentence: nltk.word_tokenize(sentence))
df["abstrak"]
```

```python
txt_stopwords = stopwords.words('indonesian')

def stopwords_removal(filtering) :
  filtering = [word for word in filtering if word not in txt_stopwords]
  return filtering

df["abstrak"] = df["abstrak"].apply(stopwords_removal)
df["abstrak"]
```

```python
import spacy

nlp = spacy.load('en_core_web_sm')

# # Contoh dokumen
# documents = df['abstrak']

# # Tokenisasi dan praproses teks
# tokenized_documents = []
# for doc in documents:
#     tokens = [token.text for token in nlp(doc) if not token.is_punct and not token.is_space]
#     tokenized_documents.append(tokens)

# Ambil isi kolom "abstrak" dari dataframe df
abstracts = df['abstrak']

# Tokenisasi dan praproses teks untuk setiap dokumen dalam kolom "abstrak"
tokenized_documents = []

for doc in abstracts:
    # Jika elemen dalam kolom "abstrak" adalah list dokumen
    # Tokenisasi dan praproses setiap dokumen dalam list
    for document in doc:
        tokens = [token.text for token in nlp(document) if not token.is_punct and not token.is_space]
        tokenized_documents.append(tokens)
```

```python
from gensim import corpora, models

# Membangun kamus dari tokenized documents
dictionary = corpora.Dictionary(tokenized_documents)

# Mengonversi tokenized documents menjadi corpus
corpus = [dictionary.doc2bow(doc) for doc in tokenized_documents]

# Membangun model LDA
lda_model = models.LdaModel(corpus, num_topics=5, id2word=dictionary, passes=15)
```

```python
# Menampilkan topik-topik
topics = lda_model.print_topics(num_words=10)
for topic in topics:
    print(topic)

# Menampilkan topik utama untuk setiap dokumen
for i, doc in enumerate(corpus):
    print(f'Dokumen {i}: {lda_model[doc]}')
```

```python
import pandas as pd

# Membuat dataframe untuk menyimpan proporsi topik dalam dokumen
df = pd.DataFrame(columns=df['abstrak'] + [f'Topik {i}' for i in range(lda_model.num_topics)])

# Mengisi dataframe dengan data proporsi topik dalam dokumen
for i, doc in enumerate(corpus):
    doc_topics = lda_model[doc]
    topic_prop = [0] * lda_model.num_topics
    for topic, prop in doc_topics:
        topic_prop[topic] = prop
    df.loc[i] = [f'Dokumen {i}'] + topic_prop

# Menampilkan tabel proporsi topik dalam dokumen
print(df)
```

```python
import pandas as pd

# Membuat dataframe untuk menyimpan proporsi kata dalam topik
df = pd.DataFrame(columns=[f'Kata {i}' for i in range(len(dictionary))])

# Mengisi dataframe dengan data proporsi kata dalam topik
for i in range(lda_model.num_topics):
    topic_words = lda_model.show_topic(i, topn=len(dictionary))
    word_prop = [0] * len(dictionary)
    for word, prop in topic_words:
        word_id = dictionary.token2id.get(word)
        if word_id is not None:
            word_prop[word_id] = prop
    df.loc[f'Topik {i}'] = word_prop

# Menampilkan tabel proporsi kata dalam topik
print(df)
```
